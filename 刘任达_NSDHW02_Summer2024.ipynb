## Ito isometry

### 1. (20pts)
Find the variance of the following random variables by applying the Ito isometry 

$$
\Eof{\left(\int_a^b \varphi_s dB_s\right)^2} = \int_a^b \Eof{\varphi_s^2}ds
$$

- (a) $X = \dint_0^t B_s^2 dB_s$. In this case, $\varphi_s = B_s^2$.
- (b) $X = \int_a^b \sqrt{t|B_t|} dB_t$, where $0 < a < b$. In this case, $\varphi_t = \sqrt{t|B_t|}$.

### <font color=blue> Solution 1(a). </font>

$B_s \in L^2(\RR) \implies B_s^4 \in L^2(\RR)$, so $\Eof{\dint_a^b B_s^4 \d s} < \infty$, by Novikov's Theorem, $X_s$ is a martingale.

Applying Ito's isometry, we get
$$
    \mathrm{Var}[X_t] = \Eof{X_t^2} = \dint_0^t \Eof{B_s^4}\d s = \int_0^t 3s^2 \d s = t^3.
$$

### <font color=blue> Solution 1(b). </font>

$\PP[|B_t| < \infty] = 1 \implies \sqrt{t|B_t|} \in L^2(\RR)$, $\Longleftrightarrow \Eof{\dint_a^b s |B_s| \d s} < \infty$, by Novikov's Theorem, $X_s$ is a martingale.

Applying Ito's isometry, we get
$$
    \mathrm{Var}[X_t] = \Eof{X_t^2} = \dint_0^t \Eof{s|B_s|}\d s = \int_0^t s (\sqrt{\frac {2s} \pi} )\d s = \frac{2\sqrt 2}{5\sqrt \pi} t^{\frac 5 2}.
$$

## Apply Ito's formula to solve SDE

### 2. (10pts)

Let $X_t$ satisfy the SDE for geometric Brownian motion

$$
dX_t = \mu(\sigma_1 X_t + \sigma_0)dt + (\sigma_1 X_t + \sigma_0)dB_t, \quad X_0 = x,
$$

where $\mu$, $\sigma_0$ and $\sigma_1 \neq 0$ are constants. 
Apply Ito's formula to $f(X_t) = \ln (\sigma_1 X_t + \sigma_0)$ then solve the SDE.

#### Note
- $f(X_t)$ is referred to as *shifted lognormal* distributed.
- Black-Scholes model can be recovered by setting $\sigma_0 = 0$.

### <font color=blue> Solution 2. </font>

Applying Ito's formula, we get

$$
\begin{equation}
    \begin{aligned}
        \d\left(\ln (\sigma_1 X_t + \sigma_0)\right) & = \frac{1}{X_t + \frac{\sigma_0}{\sigma_1}} \d X_t - \frac{1}{2(X_t + \frac{\sigma_0}{\sigma_1})^2} \d\langle X \rangle_t\\
        & = \mu \sigma_1 \d t + \sigma_1 \d B_t - \frac{\sigma_1^2}{2} \d t
    \end{aligned}
\end{equation}
$$

Integrating both sides, we obtain

$$
\begin{equation}
\begin{aligned}
    \ln (\sigma_1 X_t + \sigma_0) - \ln (\sigma_1 x + \sigma_0) & = (\mu\sigma_1 - \frac{\sigma_1^2}{2}) t + \sigma_1 B_t \\
    \Longleftrightarrow \sigma_1 X_t + \sigma_0 & = (\sigma_1 x + \sigma_0) \exp((\mu\sigma_1 - \frac{\sigma_1^2}{2}) t + \sigma_1 B_t).
\end{aligned}
\end{equation}
$$

## Relationship between SDE and PDE

### 3. (20pts)

Let $X_t$ satisfy the SDE for Ornstein-Uhlenbeck process

$$
dX_t = \lambda(m - X_t) dt + \sigma dB_t, \quad X_0 = x > 0,
$$

where $\lambda > 0$, $m$ and $\sigma$ are constants. 

- (a) Apply Ito's formula to $e^{\lambda t}X_t$ and show that the solution is given by

$$
X_t = e^{-\lambda t} x + \left(1 - e^{-\lambda t} \right) m + \sigma \int_0^t e^{-\lambda(t - s)} dW_s.
$$

- (b) Apply Feynman-Kac formula to find the moment generating function for the random variable $\int_0^T X_t dt$, i.e., the following expectation

$$
\Eof{\left.e^{-\xi\int_0^T X_s ds}\right|X_0 = x},
$$

where $\xi \in \mathbb{R}$ is a parameter.

<font color=blue>Hint</font>: To solve the resulting PDE for (b), use the ansatz 

$$
u(x, t) = e^{\alpha(t)x + \beta(t)}.
$$

### <font color=blue> Solution 3(a). </font>

Applying Ito's formula, we get
$$
\begin{equation}
\begin{aligned}
\d\left(e^{\lambda s}X_s\right) & =  \lambda e^{\lambda s} X_s\d s + e^{\lambda s}  \d X_s\\
& = \lambda e^{\lambda s} X_s \d s + e^{\lambda s}( \lambda (m - X_s) \d s + \sigma \d B_s).\\
& = \lambda m e^{\lambda s} \d s + \sigma e^{\lambda s} \d B_s

\end{aligned}
\end{equation}
$$
Integrating w.r.t. s, and divide by $e^{\lambda t}$ from both sides, we obtain
$$
X_t = e^{-\lambda t} x + \left(1 - e^{-\lambda t} \right) m + \sigma \int_0^t e^{-\lambda(t - s)} \d B_s.
$$

### <font color=blue> Solution 3(b). </font>

By Feymann-Kac formula, if a $C^2$ function $u$ satisfies:
$$
\begin{equation}
    \begin{aligned}
    u_t + \frac{\sigma^2}{2} u_{xx} + \mu u_x &= \xi x u, \quad t < T,  \\
    u(x, T) &\equiv 1. \quad \textrm{Here, } \mu(X_t, t) = \lambda (m - X_t).
    \end{aligned}
\end{equation}
$$

The solution to the PDE is given by:
$$
  u(x,t) = \E_{t,x}\left[e^{-\int_t^T \xi X_s ds}\right], \textrm{ in particular } u(x, 0) = \E[e^{-\xi\int_0^T X_s ds}|X_0 = x]
$$

Take the form $u(x, t) = \exp(\alpha(t) x + \beta(t))$, where $\alpha, \beta \in C^2$, we have $u_t = u(\alpha'x + \beta')$, $u_x = \alpha u$, $u_{xx} = \alpha^2 u$.

By plugging in the ansatz to the PDE, we obtain: 
$$
\begin{equation*} 
\begin{aligned}
  x\alpha' + \beta' + \frac{\sigma^2}{2}\alpha^2 + \lambda(m - x) \alpha  &= x(\alpha' - \lambda \alpha) + \beta' + \frac{\sigma^2}{2}\alpha^2 + \lambda m \alpha = \xi x, \\
  u(x, T) &= \exp(\alpha(T) x + \beta(T)) \equiv 1.\\
  \implies \alpha' -\lambda \alpha &= \xi, \alpha(T) = 0, \implies \alpha =  \frac{\xi}{\lambda} (e^{\lambda \tau} - 1).\ (\textrm{let } \tau = t-T)\\
  \implies \beta' = -m\xi(e^{\lambda\tau} - 1) &- \frac{\sigma^2\xi^2}{2\lambda^2} (e^{2 \lambda \tau} -2 e^{\lambda\tau}+ 1).\\
  \implies \beta = -\frac{\sigma^2 \xi^2}{4\lambda^3}(e^{2 \lambda \tau} - 1) &+ (\frac{\sigma^2 \xi^2}{2\lambda^3} - \frac{m\xi}{\lambda})(e^{\lambda \tau} - 1) + (m \xi - \frac{\sigma^2 \xi^2}{2\lambda^2})\tau
\end{aligned}
\end{equation*}
$$

Therefore, the moment generating function for the random variable is $u(x, 0) = \exp\left[(\frac{\xi}{\lambda} (e^{-\lambda T} - 1))x + -\frac{\sigma^2 \xi^2}{4\lambda^3}(e^{-2 \lambda T} - 1) + (\frac{\sigma^2 \xi^2}{2\lambda^3} - \frac{m\xi}{\lambda})(e^{-\lambda T} - 1) - (m \xi - \frac{\sigma^2 \xi^2}{2\lambda^2})T\right]$

REMARK. Inside the $\exp$ part is a quadratic polynomial of $\xi$, which suggests $\dint_0^T X_s \d s$ is a Gaussian process.

## Simulating the Black-Scholes model

### 4. (20pts)

With 100 steps within the time interval $[0,1]$, simulate 20,000 sample paths for the Black-Scholes model 

$$
dX_t = r X_t dt + \sigma X_t dB_t
$$

by the Euler-Maruyama scheme with $X_0 = 100$. The parameters are given by $r = 2\%$ and $\sigma = 25\%$.

- (a) Plot the histograms of $X_t$ for $t = 0.4, 1$.
- (b) Evaluate the expectation $e^{-rT}\Eof{g(X_T)}$ by taking sample mean from using the simulated samples in (a), where $T = 1$ and $g$ is given by 
$$
g(x) = \left\{\begin{array}{ll}
5 & \text{ if } x > 101, \\
-4 & \text{ if } x \leq 101.
\end{array}\right.
$$
Plot the histogram of $g(X_1)$. 

Note: In `python`, the function $g$ can be coded as `5*(x>101) - 4*(x<=101)`. 



```python
# as always, import required modules and functions
import numpy as np
from numpy import sqrt, log, exp
import matplotlib.pyplot as plt
import scipy.stats as ss
from scipy.stats import norm
import seaborn as sns
```

```python
#(4a)
# set random seed
np.random.seed(1609)

# parameters
x0, sigma, r, T = 100, 0.25, 0.02, 1 
n_sim, n_steps = 20_000, 100
dt = T/n_steps

# initialize X
X = np.ones([n_sim, n_steps+1])*x0

# Euler-Maruyama scheme
for i in range(n_steps):
    db = norm.rvs(size=n_sim)
    db = db - db.mean()
    db = db/db.std()
    X[:,i+1] = X[:,i] + r*X[:,i]*dt + sigma*X[:,i]*np.sqrt(dt)*db

# plot
plt.figure(figsize=(12, 8))
t = np.arange(0, T+dt, dt)

# histograms at different times
plt.subplot(2, 2, 1)
time = 0.4
sns.histplot(X[:,t==time], bins=30, stat='density', label=f'$t={time}$', element='step')
sns.histplot(X[:,-1], bins=30, stat='density', label=f'$t={T}$', color='lightgreen', element='step')
plt.xlabel('$X_t$')
plt.ylabel('Density')
plt.title('Density of $X_t$')
plt.legend()

# time evolution of mean
plt.subplot(2, 2, 2)
plt.plot(t, X.mean(axis=0))
plt.xlabel('$t$')
plt.ylabel('sample mean')
plt.title('Sample Mean of $X_t$')

# time evolution of variance
plt.subplot(2, 2, 3)
plt.plot(t, X.var(axis=0))
plt.xlabel('$t$')
plt.ylabel('sample variance')
plt.title('Sample Variance of $X_t$')

plt.tight_layout()
plt.show();
```

```python
#(4b)
g = lambda x: 5*(x>101) - 4*(x<=101)

exp(-r*T)*g(X[:,-1]).mean()
```
